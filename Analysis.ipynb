```python
f = open('national-average.csv')

lines = []
for line in f:
    lines.append(line.strip())


print(lines)
```

```python
rows = []
for line in lines:
    row = line.strip().split(',')
    rows.append(row)
print(rows)
```

```python
for row in rows[1:]:
    if int(row[1]) >= 136:
        print("High prices:", row)
```

```python
for row in rows[1:]:
    if int(row[1]) <= 136:
        print("Low prices:", row)
```
